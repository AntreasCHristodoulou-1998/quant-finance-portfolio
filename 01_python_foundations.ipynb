# Phase 1 - Python Foundations for Finance
## Lesson 1: Variables and Basic Calculations

```python
# --- Βασικές μεταβλητές σε Finance context ---

stock_price = 150.25        # τιμή μετοχής σε $
shares = 100                # αριθμός μετοχών
annual_return = 0.08        # 8% ετήσια απόδοση
years = 5                   # χρονικός ορίζοντας

# Αξία χαρτοφυλακίου
portfolio_value = stock_price * shares
print("Αξία χαρτοφυλακίου:", portfolio_value, "$")

# Future Value — τι θα αξίζει σε 5 χρόνια με 8% return
future_value = portfolio_value * (1 + annual_return) ** years
print("Future Value σε 5 χρόνια:", round(future_value, 2), "$")
```

```python
# --- Lists και Loops με Finance context ---

# Τιμές μετοχής για 5 μέρες
prices = [150.25, 152.10, 149.80, 153.40, 155.00]

# Υπολογισμός daily returns
for i in range(1, len(prices)):
    daily_return = (prices[i] - prices[i-1]) / prices[i-1] * 100
    print(f"Μέρα {i}: {round(daily_return, 3)}%")
```

```python
# --- Το ίδιο με NumPy --- πολύ πιο γρήγορα!

import numpy as np

prices = np.array([150.25, 152.10, 149.80, 153.40, 155.00])

# Υπολογισμός όλων των returns σε ΜΙΑ γραμμή
returns = np.diff(prices) / prices[:-1] * 100

print("Daily Returns (%):", np.round(returns, 3))
print("Μέση απόδοση:", round(np.mean(returns), 3), "%")
print("Μέγιστη απόδοση:", round(np.max(returns), 3), "%")
print("Ελάχιστη απόδοση:", round(np.min(returns), 3), "%")
```

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
prices = [150.25, 152.10, 149.80, 153.40, 155.00]

plt.figure(figsize=(8, 4))
plt.plot(days, prices, marker='o', color='blue', linewidth=2)
plt.title("Stock Price - 5 Days")
plt.xlabel("Day")
plt.ylabel("Price ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
